# Gold Analysis

```python
# dependencies
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# store file path
path = os.path.join('gold_data.csv')

# store in dataframe
df = pd.read_csv(path)

df.head()
```

```python
# plot
x = df['Week']
y1 = df['USD (PM)']
fig, ax1 = plt.subplots()
ax1.plot(x, y1, 'g-')
ax1.set_xlabel('Week')
ax1.set_ylabel('Price ($)', color='g')
plt.title('Price of Gold Over 2017')
fig.set_size_inches(18.5, 10.5)
plt.savefig('gold_price.png')
plt.show()
```

```python
# add empty column
df['Percentage Change in Price'] = ''

# Calculate percentage changes
df['Percentage Change in Price'] = df['USD (PM)'].pct_change()

df.to_csv('gold_results.csv')

df.head()
```

```python
# plot
x = df['Week']
y1 = df['Percentage Change in Price']
fig, ax1 = plt.subplots()
ax1.plot(x, y1, 'g-')
ax1.set_xlabel('Week')
ax1.set_ylabel('% Change in Price', color='g')
plt.title('% Change in Price of Gold Over 2017')
fig.set_size_inches(18.5, 10.5)
plt.savefig('gold_percent_changes.png')
plt.show()
```
